```python
import pandas as pd
podatki = pd.read_csv('stevilke.csv')

pd.options.display.max_rows = 20
podatki['stevilke']
stevilke = podatki.groupby('stevilke')
to = stevilke.count()
to.sort_values('datum', ascending = False).head(10)
```

```python
euro = pd.read_csv('stevilke_euro.csv')
stevilke_euro = euro.groupby('stevilke')
tisto = stevilke_euro.count()
tisto.sort_values('datum', ascending = False)
```
